```python
#Faça um programa que leia um ângulo qualquer e mostre na tela o valor de
#Seno, Cosseno e Tangente desse ângulo

print('Seno, Cosseno e Tangente \n')

from math import cos, sin, tan, radians

ang = int(input('Digite um ângulo: '))

ang_rad = radians(ang)

seno = sin(ang_rad)
cosseno = cos(ang_rad)
tangente = tan(ang_rad)

print(f'O seno de {ang}° é: {seno:.2f}.\n'
      f'O cosseno de {ang}° é {cosseno:.2f}.\n'
      f'E a tangente de {ang}° é {tangente:.2f}.')
```
