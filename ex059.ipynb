```python
# Crie um programa que leia dois valores e mostre um menu na tela:
[1] somar
[2] multiplicar
[3] maior
[4] novos números
[5] sair do programa
# Seu programa deverá realizar a operação solicitada em casa caso.
```

```python
num1 = int(input('Digite um número: '))
num2 = int(input('Digite outro número: '))

flag = False

while not flag:

 print('''
[1] somar
[2] multiplicar
[3] maior
[4] novos números
[5] sair do programa
''')

 opcao = int(input('Escolha uma opção: '))

 if opcao == 1:
   print(num1+num2)

 if opcao == 2:
   print(num1*num2)

 if opcao == 3:
   if num1 > num2:
    print(num1)
   if num2 > num1:
    print(num2)
   if num1 == num2:
    print('Os números são iguais!')

 if opcao == 4:
   num3 = int(input('Digite um novo número: '))
   num4 = int(input('Digite outro novo número: '))
   num1 = num3
   num2 = num4

 if opcao == 5:
   flag = True
```
