```python
# FaÃ§a um programa que leia o peso de cinco pessoas.
# No final, mostre qual foi o maior e o menor peso lidos.
```

```python
maior = 0
menor = 0

for p in range (1, 3):
  peso = float(input('Digite aqui o seu peso em KG: '))
  if p == 1:
    maior += peso
    menor += peso

  else:
    if peso > maior:
      maior = peso
    if peso < menor:
      menor = peso
print(f'O maior peso foi de {maior}')
print(f'E o menor foi de {menor}')
```
