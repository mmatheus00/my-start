```python
# Escreva um programa que leia a velocidade de um carro.
# Se ele ultrapassar 80km/h, mostre uma mensagem dizendo que ele foi multado.
# A multa vai custar R$7.00 por cada Km acima do limite.

print('~~'*25)
print('Radar a frente! Velocidade máxima da via: 80 km/h')
print('~~'*25)

km_max_via = 80
km_atual = int(input('Qual sua velocidade atual?: '))
km_multa = (km_atual - km_max_via)

if km_atual >km_max_via:
  print(f'Você estava acima do limite da via quando passou pelo radar. Você foi multado em R${km_multa*7}!\n')
else:
  print('Dirija com cuidado!\n')

print('----------FIM----------')

```
