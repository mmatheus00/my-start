```python
# Desenvolva uma lógica que leia o peso e altura de uma pessoa, calcule seu IMC e mostreu seu status, de acordo com a tabela abaixo:

# Abaixo de 18.5: Abaixo do Peso
# Entre 18.5 e 25: Peso Ideal
# de 25 até 30: Sobrepeso
# de 30 até 40: Obesidade
# Acima de 40: Obesidade Mórbida

# Fórmula IMC = kg / (altura * altura)

kilo = float(input('Digite aqui o seu peso: '))
altura = float(input('Digite aqui a sua altura: '))
imc = kilo / (altura * 2)

if imc < 18.5:
  print(f'O seu IMC é de {imc:.2f}, você está Abaixo do Peso Ideal!')
elif 18.5 <= imc <= 25:
  print(f'O seu IMC é de {imc:.2f}, você está com o Peso Ideal!')
elif 25 <= imc <= 30:
  print(f'O seu IMC é de {imc:.2f}, você está com Sobrepeso!')
elif 30 <= imc <= 40:
  print(f'O seu IMC é de {imc:.2f}, você está com Obesidade!')
elif imc > 40:
  print(f'O seu IMC é de {imc:.2f}, você está com Obesidade Mórbida!')

print('----------------------FIM----------------------')
```
