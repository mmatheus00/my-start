```python
# Faca um programa que jogue impar ou par com o computador.
# O jogo so sera interrompido quando o jogador PERDER,
# Mostrando o total de vitorias consecutivas que ele conquistou no final do jogo.
```

```python
import random

computador = streak = 0

print('~'*45)
print(' '*7,'Vamos Jogar Impar ou Par!')
print('~'*45)
print('\n')

while True:
  user = int(input('Escolha seu numero: '))
  escolha = str(input('Impar ou Par? [I/P] ')).upper().strip()[0]
  print('\n')
  computador = random.randint(0, 10)
  soma = computador + user
  par = soma % 2

  if escolha != 'I' and escolha != 'P':
    print('Entrada Invalida!')

  if par == 1 and escolha == 'I':
    streak +=1
    print(f'O computador jogou {computador} e voce jogou {user} a soma deu {soma}!')
    print('\n')
    print('=-'*23)
    print(' '*10,'DEU IMPAR. VOCE VENCEU!')
    print('=-'*23)
    print('\n')
    print('O computador quer uma REVANCHE...')
    print('~'*45)
    print('\n')

  elif par == 0 and escolha == 'P':
    streak +=1
    print(f'O computador jogou {computador} e voce jogou {user} a soma deu {soma}!')
    print('\n')
    print('=-'*23)
    print(' '*10,'DEU PAR. VOCE VENCEU!')
    print('=-'*23)
    print('\n')
    print('O computador quer uma REVANCHE...')
    print('~'*45)
    print('\n')

  else:
    print(f'O computador jogou {computador} e voce jogou {user} a soma deu {soma}!')
    print('\n')
    if par != 0 and escolha == 'P':
      print('=-'*23)
      print(' '*12,f'DEU IMPAR. VOCE PERDEU!')
      print('=-'*23)
    elif par == 0 and escolha == 'I':
      print('=-'*23)
      print(' '*12,f'DEU PAR. VOCE PERDEU!')
      print('=-'*23)
    print('\n')
    print(f'Voce ganhou do computador {streak} vezes!')
    print('\n')
    print('~'*45)
    break
```

Correção Guanabara

```python
from random import randint

v = 0
while True:
    jogador = int(input('Diga um valor: '))
    computador = randint(0, 10)
    total = jogador + computador
    tipo = ' '
    while tipo not in 'PI':
        tipo = str(input('Par ou Ímpar? [P/I] ')).strip().upper()[0]
    print(f'Você jogou {jogador} e o computador {computador}. Total de {total}', end=' ')
    print('DEU PAR' if total % 2 == 0 else 'DEU ÍMPAR')
    if tipo == 'P':
        if total % 2 == 0:
            print('Você VENCEU!')
            v += 1
        else:
            print('Você PERDEU!')
            break
    elif tipo == 'I':
        if total % 2 == 1:
            print('Você VENCEU!')
            v += 1
        else:
            print('Você PERDEU!')
            break
    print('Vamos jogar novamente...')
print(f'GAME OVER! Você venceu {v} vezes.')
```
