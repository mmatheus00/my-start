```python
# Desenvolva um programa que leia o primeiro termo e a razão de uma progressão aritmética.
# No final, mostre os 10 primeiros termos dessa progressão.
```

```python
termo = int(input('Digite aqui o primeiro termo: '))
razao = int(input('Digite aqui a razão: '))

seg_termo = (termo + (10 - 1) * razao) + razao
print('\n')

print(f'Esses são os dez primeiros termos dessa progressão aritmética: ', '\n')

for pa in range (termo, seg_termo, razao):
  print(pa, end=' ')
```
