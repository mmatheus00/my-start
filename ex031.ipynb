```python
# Desenvolva um programa que pergunte a distância de uma viagem em Km.
# Calcule o preço da passagem, cobrando R$0.50 por Km para viagens até 200km,
# e R$0.45 para viagens mais longas.

city = str(input('Digite o nome da cidade que você vai visitar: ')).strip().title()
travel = int(input('Qual a distância em Km da sua cidade até seu destino? '))

if travel <=200:
  print(f'Para uma viagem de {travel}km, sua passagem vai custar R${travel*0.5:.2f}!')
else:
  print(f'Para uma viagem de {travel}km, sua passagem vai custar R${travel*0.45:.2f}!')
```
