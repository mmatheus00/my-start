```python
# Faça um programa que leia o ano de nascimento de um jovem e informe, de acordo com sua idade:
# Se ele ainda vai se alistar ao serviço militar
# Se é a hora de se alistar
# Se ja passou do tempo de se alistar
# Seu programa também deve mostrar o tempo que falta ou ja passou do prazo.


ano_input = (int(input('Digite o ano de seu nascimento: ')))
if ano_input > 2007:
  print('Ainda não chegou seu tempo de alistamento.')
  print(f'\33[1;33mfaltam {18-(2025 - ano_input)} ano(s) para seu alistamento!\33[0m')
elif ano_input == 2007:
  print('\33[1;31mCHEGOU A SUA HORA DE SE ALISTAR!\33[0m')
elif ano_input < 2007:
  print('Ja passou o sufoco, fica suave.')
  print(f'\33[1;32mSeu alistamento foi há {(2025 - ano_input)-18} ano(s)!\33[0m')
print('-------------FIM-------------')
```
