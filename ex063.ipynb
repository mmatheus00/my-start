```python
# escreva um programa que leia um numero N inteiro qualquer e mostre na tela os N primeiros elementos de uma sequencia de fibonacci.
# 0 -> 1 -> 1 -> 2 -> 3 -> 5 -> 8

# Em termos matemáticos, a sequência é definida pela fórmula Fn = fn-1 + Fn-2, sendo o  primeiro termo F1= 1 e os valores iniciais F1 = 1, F2 =1.
```

```python
n = int(input('Digite o numero de termos de fibonacci voce deseja ver: '))

t1 = 0
t2 = 1
contador = 3
print(f'{t1} >>> {t2}', end=' >> ')

while contador <= n:
  t3 = t1 + t2
  print(t3, end=' >> ')
  t1 = t2
  t2 = t3
  contador +=1
print('FIM')
```
