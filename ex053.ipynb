```python
# Crie um programa que leia uma frase qualquer e diga se ela é um Palíndromo, desconsiderando os espaços.
```

Minha resolucao

```python
frase = str(input('Digite uma frase: ')).strip().lower().split()
frase = ''.join(frase)

frase_2 = (''.join(frase[::-1]))

if frase == frase_2:
  print(f'{frase} e igual a {frase_2}, sendo assim, e um Palíndromo!')
else:
  print(f'{frase} nao e igual a {frase_2}, sendo assim, nao e um Palíndromo!')
```

Resolucao do Guanabara

```python
frase = str(input('Digite uma frase: ')).strip().lower().split()
frase = ''.join(frase)

frase_2 = ''

for l in range(len(frase) -1, -1, -1):
  frase_2 += frase[l]

if frase == frase_2:
  print(f'{frase} e igual a {frase_2}, sendo assim, e um Palíndromo!')
else:
  print(f'{frase} nao e igual a {frase_2}, sendo assim, nao e um Palíndromo!')
```
