```python
# Crie um programa que leia varios numeros inteiros pelo teclado.
# O programa so vai parar quando o usuario digitar o numero 999 (condicao de parada).
# No final, mostre quantos numeros foram digitados e qual foi a soma entre eles (desconsiderando a condicao de parada)
```

```python
user = cont = acumm = 0

print('''Digite alguns numeros para para descobrir qual a soma entre eles.

Para sair do programa digite [ 999 ]''')
print('\n')

while user != 999:
  user = int(input('Digite um numero inteiro: '))
  if user != 999:
    acumm += user
    cont += 1
print('\n')
print(f'Voce digitou {cont} numeros e a soma de todos eles e igual a {acumm}! \n')
print('[Fim do programa]')
```
