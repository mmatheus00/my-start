```python
# Escreva um programa que leia dois números inteiros e compare-os, mostrando na tela uma mensagem:

# O primeiro valor é maior
# O segundo valor é maior
# não existe valor maior, os dois são iguais.

print('Analisador de Números')

n1 = int(input('Digite o primeiro número: '))
n2 = int(input('Digite o segundo núemro: '))

if n1 > n2:
  print(f'O número \33[1;32m{n1}\33[0m é maior que o número \33[1;31m{n2}\33[0m!')
elif n2 > n1:
  print(f'O número \33[1;32m{n2}\33[0m é maior que o número \33[1;31m{n1}\33[0m!')
elif n1 == n2 and n2 == n1:
  print(f'Os números \33[1;33m{n1}\33[0m e \33[1;33m{n2}\33[0m são iguais!')

print('-------------FIM-------------')
```
