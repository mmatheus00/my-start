```python
# Faça um programa que leia três números e mostre qual o maior e qual é o menor

num = float(input('Digite um número qualquer: '))
num2 = float(input('Digite outro número qualquer: '))
num3 = float(input('Digite um terceiro número qualquer: '))

if num > num2 and num2 > num3:
  print(f'{num} é o número maior!')
elif num < num2 and num2 < num3:
  print(f'{num3} é o número maior!')
elif num < num2 and num2 > num3:
  print(f'{num2} é o número maior!')
```
