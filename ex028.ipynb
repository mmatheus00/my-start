```python
# Escreva um programa que faça o computador pensar em um numero inteiro de 0 a 5.
# e peça para o usuário tentar descobrir qual foi o número escolhido pelo computador
# O programa deverá escrever na tela se o usuário venceu ou perdeu.


import random

number = random.randint(0, 5)

print('~~~~'*16)
print('Pensei em um número entre 0 e 5. Eu Dúvido você advinhar! >:D')
print('~~~~'*16)

print('\n')
guess_number = int(input('Advinhe meu número: '))
print('\n')

if number == guess_number:
  print(f'NOSSA! Você é melhor do que eu pensava. VOCÊ ADVINHOU O MEU NÚMERO {number}!! <:O\n')

elif guess_number >5:
  print('NÚMERO INVÁLIDO! Só valem números de 0 a 5!! -.-\n')

else:
  print(f'SABIA! Você nunca advinharia que eu pensei no número {number}!! >:D\n')

print('----------FIM----------')
```
