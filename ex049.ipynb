```python
# refaça o desafio 009, mostrando a tabuada de um número que o usuário escolher
# só que agora utilizando o laço for
```

```python
print('~'*18)
print('Hora da tabuada v2')
print('~'*18)
print('\n')
num = int(input('Digite um número inteiro: '))
print('\n')
print('-'*18)
print(f'Tabuada de {num}:')
print('-'*18)

for n in range (1, 11):
  print(f'{num} * {n} = {num*n}')
print('-'*18)
```
