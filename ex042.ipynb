```python
# refaça o desafio 035 dos triângulos, acrescento o recurso de mostrar que tipo de triângulo será formado:

# Equilátero = todos os lados iguais
# Isósceles = dois lados iguais
# Escaleno = todos os lados diferentes

pri = int(input('Digite o comprimento de uma reta: '))
seg = int(input('Digite o comprimento de outra reta: '))
ter = int(input('Digite o comprimento da última reta: '))

if pri == seg and seg == ter and ter == pri:
  print('Com esses valores nós formamos um triângulo Equilátero!')
elif pri == seg or seg == ter or ter == pri:
  print('Com esses valores nós formamos um triângulo Isósceles!')
elif pri != seg and seg != ter and ter != pri:
  print('Com esses valores nós formamos um triângulo Escaleno!')

print('-----------------FIM-----------------')
```
