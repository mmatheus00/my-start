```python
# Crie um programa que leia a data de nascimento de sete pessoas.
# No final, mostre quantas pessoas ainda não atingiram a maioridade
# e quantas ja são maiores.
```

Minha resolucao

```python
ano = 2025
for a in range(1, 8):
  nasc = int(input('Digite seu ano de nascimento: '))
  tempo = ano - nasc
  if tempo > 18:
    print(f'Você já é de Maior a {tempo - 18} anos!')
  elif tempo == 18:
    print(f'Você acabou de ficar de Maior!')
  else:
    print(f'Faltam {18 - tempo} anos para você ser de Maior!')
```

Resolucao Guanabara

```python
from datetime import date
ano = date.today().year
cont = 0
cont2 = 0
for a in range(1, 4):
  nasc = int(input('Em que ano voce nasceu?: '))
  tempo = (ano - nasc)
  if tempo >= 18:
    cont += 1
  else:
    cont2 += 1
print(f'Nos temos {cont} pessoas maiores de idade!')
print(f'Nos temos {cont2} pessoas menores de idade!')
```
