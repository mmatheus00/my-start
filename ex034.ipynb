```python
# Escreva um programa que pergunta o salário do funcionários e calcule seu aumento.
# Para salários superiores a R$1250.00, calcule um aumento de 10%.
# Para os inferiores ou iguais o aumento é de 15%

sal_user = int(input('Digite aqui o seu salário em R$: '))
sal_base = 1250
if sal_user > sal_base:
  print(f'PARABÉNS! Você recebeu um aumento. Agora seu salário é de R${sal_user*1.10:.2f}')
elif sal_user <= sal_base:
  print(f'PARABÉNS! Você recebeu um aumento. Agora seru salário é de R${sal_user*1.15:.2f}')
```
