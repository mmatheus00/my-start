```python
# Crie um programa que leia varios numeros inteiros pelo teclado
# No final da execucao, mostre a media entre todos os valores e qual foi o maior e o menor valor lido.
# O programa deve perguntar ao usuario se ele quer ou nao continuar a digitar valores.
```

```python
acumm = cont = 0
user = 1
numb = []

while user == 1:
  num = int(input('Digite um numero: '))
  print('\n')
  numb.append(num)
  acumm += num
  cont += 1
  print('''Deseja digitar mais numeros?

[ 1 ] Sim!
[ 2 ] Nao! Sair do programa.
  ''')
  user = int(input('Digite sua opcao: '))

  while user != 1 and user != 2:
      print('Entrada Invalida')
      print('''Deseja digitar mais numeros?

[ 1 ] Sim!
[ 2 ] Nao! Sair do programa.
      ''')
      user = int(input('Digite sua opcao: '))
if user == 2:
  media = (acumm/cont)
  print(f'''
Total digitado >> {cont} numeros
Media dos numeros >> {media:.1f}
Maior numero >> {max(numb)}
Menor numero >> {min(numb)}''')
  print('>> FIM <<')
```

```python
resp = 'S'
soma = quant = média = maior = menor = 0
while resp in 'Ss':
    núm = int(input('Digite um número: '))
    soma += núm
    quant += 1
    if quant == 1:
        maior = menor = núm
    else:
        if núm > maior:
            maior = núm
        if núm < menor:
            menor = núm
    resp = str(input('Quer continuar? [S/N] ')).upper().strip()[0]
média = soma / quant
print(f'Você digitou {quant} números e a média foi {média}')
print(f'O maior valor foi {maior} e o menor foi {menor}')
```
