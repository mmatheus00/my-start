```python
# Desenvolva um programa que leia o nome, idade e sexo de 4 pessoas.
# no final do programa mostre:

# A media de idade do grupo
# Qual e o nome do homem mais velho
# Quantas mulheres tem menos de 20 anos
```

```python
maioridade = 0
nomevelho = ' '
contar = 0
acumular = 0

for dados in range (1, 5):
  nome = str(input('Digite seu nome: ')).strip().lower()
  idade = int(input('Digite sua idade: '))
  sexo = str(input('Digite seu sexo [M/F]: ')).strip().upper()

  if sexo =='M' and dados == 1:
    maioridade += idade
    nomevelho += nome
  else:
    if sexo == 'M' and idade > maioridade:
      maioridade = idade
      nomevelho = nome

  if sexo == 'F' and idade < 20:
    acumular += 1

  contar += idade

media = (contar / dados)

print(f'A media de idade das pessoas desse grupo e de {media} anos!')
print(f'O homem mais velho tem {maioridade} anos e seu nome e {nomevelho}')
print(f'No grupo temos {acumular} mulheres menores de 20 anos!')
```
