```python
# Melhora o jogo do ex028 onde o computador vai pensar em um numero entre 0 e 10.
# Só que agora o jogador vai tentar advinhar até acertar, mostrando no final quantos palpites foram necessários para vencer
```

```python
from random import randint

number = randint(0, 11)
count = 1
flag = False

print('Pensei em um número entre 0 e 10... você consegue advinhar qual é?')

while not flag: # Enquanto flag( leia como False) não for True:
  guess = int(input('Digite um número: '))
  if guess == number:
    flag = True # O estado precisa ser alterado.
    print(f'Você acertou em {count} tentativas!')
  elif guess > 10:
    print('Entrada Inválida! Só aceitamos números entre 0 e 10.')
  else:
    print('Errou! Tente novamente.')
  count += 1
print('Fim')
```
