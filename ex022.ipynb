```python
# Crie um programa que leia o nome completo de uma pessoa e mostre:

  # O nome com todas as letras maiusculas e minusculas
  # Quantas letras ao todo (Sem considerar espaços)
  # Quantas letras tem o primeiro nome

print('Analisador de Nomes\n')


# Caracteres maiusculos e minusculos
name = str(input('Digite seu nome completo: ')).strip()
print(f'\nSeu nome em Maiusculo: {name.upper()}')
print(f'Seu nome em minusculo: {name.lower()}')

# Quantas letras ao todo sem considerar espaços
dividido = name.split()
conectado = "".join(dividido)
print(f'Seu nome completo tem: {len(conectado)} letras')

# Quantas letras tem o primeiro nome
print(f'Seu primeiro nome tem: {len(dividido[0])} letras')
```
