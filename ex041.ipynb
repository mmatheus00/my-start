```python
# A Confederação Nacional de Natação precisa de um programa que leia o ano de nascimento do atleta e mostre sua categoria, de acordo com a idade:

# Até 9 anos = MIRIM
# Até 14 anos = INFANTIL
# Até 19 anos = JUNIOR
# Até 20 anos = SÊNIOR
# Acima = MASTER


ano = int(input('Digite seu ano de nascimento: '))
idade = 2025 - ano

if idade <= 9:
  print('Sua categoria é: MIRIM!')
elif idade <= 14:
  print('Sua categoria é: INFANTIL!')
elif idade <= 19:
  print('Sua categoria é: JUNIOR!')
elif idade <= 20:
  print('Sua categoria é: SÊNIOR!')
elif idade > 20:
  print('Sua categoria é: MASTER!')
print('--------------FIM--------------')
```
