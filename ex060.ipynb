```python
# Faça um programa que leia um número qualquer e mostre seu fatorial.
```

```python
from math import factorial
number = int(input('Digite um número qualquer: '))
print(factorial(number))
```

```python
number = int(input('Digite um número qualquer: '))
fact = 1

for f in range (1, number+1):
  fact = fact * f
print(fact)
```

```python
number = int(input('Digite um número qualquer: '))
fat = 1
cont = 1

while cont <= number:
  fat*=cont
  cont+=1
print(fat)
```
