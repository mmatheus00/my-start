```python
# melhore o desafio 061, perguntando para o usuario se ele quer mostrar mais alguns termos.
# O programa encerra quando ele disser que quer mostrar 0 termos.
```

```python
termo = int(input('Digite aqui o primeiro termo: '))
razao = int(input('Digite aqui a razão: '))
print('\n')

pa = 0
contador = 1
tot = 0
mais = 10

print('Aqui estao os 10 primeiros termos da PA: ')
while mais != 0:
  tot += mais
  while contador <= tot:
    pa = termo + (contador-1) * razao
    contador+=1
    print(pa, end=" >> ")
  print('PAUSA')
  mais = int(input('\nQuantos termos voce deseja adicionar?: '))
print('FIM')
```
