```python
# Desenvolva um programa que leia o comprimento de três retas e
# diga ao usuário se elas podem ou não formar um triângulo.

#Triângulo: é preciso que a soma de dois lados quaisquer seja maior que o terceiro lado.

pri = int(input('Digite o comprimento de uma reta: '))
seg = int(input('Digite o comprimento de outra reta: '))
ter = int(input('Digite o comprimento da última reta: '))

if pri <= seg + ter and seg <= pri + ter and ter <= pri + seg:
  print('Nós conseguimos formar um triângulo!')
else:
  print('Com esses valores, não conseguimos formar um triângulo!')

#if pri + seg >= ter:
```
