```python
# Faça um programa que leia qualquer número de 0 a 9999 e mostre na tela cada um dos digitos separados.
# Milhares, Centenas, Dezenas e Unidades.

# Não sei da onde saiu isso, Guanabara tava chapando.

print('Analisador de Números\n')

number = (int(input('Digite um número entre 0 e 9999: ')))
u = number // 1 % 10
d = number // 10 % 10
c = number // 100 % 10
m = number // 1000 % 10

print(f'Seu número tem:\n{u} unidades')
print(f'{d} dezenas')
print(f'{c} centenas e')
print(f'{m} milhares.')
```
