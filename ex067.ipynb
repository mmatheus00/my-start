```python
# Faca um programa que mostre a tabuada de varios numeros, um de cada vez, para cada valor digitado pelo usuario.
# O programa sera interrompido quando o numero solicitado for negativo.
```

```python
contador = 0

print('~'*18)
print('Hora da tabuada v3')
print('~'*18)
print('\n')

while True:
  print('-'*45)
  user_num = (int(input('Digite um numero para ver sua tabuada: ')))
  print('-'*45)
  if user_num < 0:
    print('Programa Encerrado!')
    break
  while contador <=11:
    print(f'{user_num} x {contador} = {user_num*contador}')
    contador += 1
print('-'*45)
```

Correção Guanabara

```python
while True:
    n = int(input('Quer ver a tabuada de qual valor? '))
    print('-' * 30)
    if n < 0:
        break
    for c in range(1, 11):
        print(f'{n} x {c} = {n*c}')
    print('-' * 30)
print('PROGRAMA TABUADA ENCERRADO. Volte sempre!')
```
