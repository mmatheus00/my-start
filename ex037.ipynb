```python
# Escreva um programa que leia um número inteiro qualquer e peça para o usuário escolher qual será a base de conversão.
# 1 = bin
# 2 = oct
# 3 = hex

num = int(input('Digite um número qualquer: '))
print('1 - Binário\n'
      '2 - Octadecimal\n'
      '3 - Hexadecimal')
choose = int(input('Escolha uma das formas de conversão digitando 1, 2 ou 3!: '))

if choose == 1:
  print(bin(num))
elif choose == 2:
  print(oct(num))
elif choose == 3:
  print(hex(num))
else:
  print('Digite um valor válido. As opções são: 1, 2 e 3!')

print('------------FIM------------')

#else:
  #print('DADOS DIVERGENTES! Por favor, escolha um número entre 1 e 3.')
```
