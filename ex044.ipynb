```python
# Elabore um programa que calcule o valor a ser pago por um produto, considerando o seu preço normal e condição de pagamento.

# À vista dinheiro/cheque: 10% de desconto.
# À vista no cartão: 5% de desconto.
# Em até 2x no cartão: preço normal.
# 3x ou mais no cartão: 20% de juros.

exit = 'Tecle para sair.'
id = 'Fone de Ouvido'
id2 = 'Cadeira Ergonomica'
id3 = 'Notebook com intel i5'
preco1 = 54.99
preco2 = 549.99
preco3 = 3299.99

print('Bem vindo a feira do rolo! Escolha um dos produtos abaixo para comprar:')
print('\n')
print(f'1 - {id}\n'
      f'2 - {id2}\n'
      f'3 - {id3}\n'
      f'0 - {exit}')
print('\n')
user_choose = (int(input('Qual você gostaria de comprar hoje?: ')))
print('\n')

if user_choose == 0:
  print('Volte sempre!')
elif user_choose == 1:
  print(f'O preço do item: {id} é R${preco1}\n'
        f'Pagando Á vista no dinheiro ou cheque, ele sai por R${preco1*(1-0.1):.2f}\n'
        f'Pagando À vista no cartão, ele sai por R${preco1*(1-0.05):.2f}\n'
        f'Pagando em até 2x no cartão, ele sai por R${preco1}. Em 3x ou mais no cartão, R${preco1*(1+0.2):.2f}'
        )
elif user_choose == 2:
  print(f'O preço do item: {id2} é R${preco2}\n'
        f'Pagando Á vista no dinheiro ou cheque, ele sai por R${preco2*(1-0.1):.2f}\n'
        f'Pagando À vista no cartão, ele sai por R${preco2*(1-0.05):.2f}\n'
        f'Pagando em até 2x no cartão, ele sai por R${preco2}. Em 3x ou mais no cartão, R${preco2*(1+0.2):.2f}'
        )
elif user_choose == 3:
  print(f'O preço do item: {id3} é R${preco3}\n'
        f'Pagando Á vista no dinheiro ou cheque, ele sai por R${preco3*(1-0.1):.2f}\n'
        f'Pagando À vista no cartão, ele sai por R${preco3*(1-0.05):.2f}\n'
        f'Pagando em até 2x no cartão, ele sai por R${preco3}. Em 3x ou mais no cartão, R${preco3*(1+0.2):.2f}'
        )
print('\n')
print('-------------------------------------------------FIM-------------------------------------------------')
```
